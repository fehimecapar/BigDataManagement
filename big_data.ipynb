# TaskI-a
Write a list comprehension that takes a number n and returns a list of even numbers, using a lambda function. 

```python
n = int(input("input a number: "))

even_nums = [x for x in range(n) if (lambda x: x % 2 == 0) (x)]

print("even numbers: ", even_nums)
```

# TaskI-b

First, write a function that takes a length in inches and returns a length in cm. Given a list l with lengths in inches: l = [4,4.5,5,5.5;6,7]. Write a list comprehension that takes l and returns a list with all values converted to cm using map.

[convert from inches to centimeters](https://www.codechef.com/practice/course/python-coding-challenges/PYTHONCHC01/problems/PYTHONCH09)


```python
def convert_to_centimeters(l):

    centimeters = [x for x in map(lambda x: x * 2.54, l)]
    
    return centimeters

l = [4,4.5,5,5.5,6,7]

print(convert_to_centimeters(l))
```

# TaskI-c
Write a list comprehension, which filters the list l from (b) by returning only sizes between 4 and 6 inches. Use filter for this!

```python
filtered_nums = [x for x in filter(lambda x: 4 <= x <= 6, l)]

print(filtered_nums)
```

# TaskI-d
Write a list comprehension that reduces the list l by summing up all lengths.

Hint: For using the reduce function, you need to import it first by adding the line: from functools import reduce

```python
from functools import reduce # import functools for reduce function

total = reduce(lambda x,y: x+y, l)

print("sum value of l list: ", total)
```
