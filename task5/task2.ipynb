*How is a heavy hitter in a graph defined?*

In graph theory, a "heavy hitter" is a node that stands out due to its significantly higher metrics compared to other nodes in the graph. This can be defined primarily by the node's degree, which indicates the number of connections (edges) it has to other nodes, making it a central part of the network. Alternatively, heavy hitters may be identified through various centrality measures, such as betweenness or closeness centrality, which assess the node's influence and connectivity within the network. In specific contexts, heavy hitters might also refer to nodes that manage a large volume of activity or traffic, highlighting their importance in the overall structure and function of the graph. Thus, heavy hitters play a crucial role in network analysis, aiding in the identification of key players and relationships within complex systems.
